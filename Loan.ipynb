```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import csv
```

```python
df=pd.read_csv("loan_train.csv")
df1=pd.read_csv("loan_test.csv")
combine=[df,df1]
```

```python
#types of features
df.dtypes
#drop unrequired things

df.drop('Loan_ID',axis=1,inplace=True)
#for counting values in the data frame
df['Gender'].value_counts()
#for couting null values in data frame
sum(pd.isnull(df['Gender']))
#for filling the missing values
df['Gender'].fillna('Male',inplace=True)
#to fill the values with the required strings
y_map={'Male':0,'Female':1}
df['Gender']=[y_map[item] for item in df['Gender']]
```

```python
#for getting the total sum of values in the particular feature
df['Married'].value_counts().sum()
df['Married'].fillna('Yes',inplace=True)
y_map={'Yes':1,'No':0}
df['Married']=[y_map[item] for item in df['Married']]
```

```python
df['Dependents'].fillna('0',inplace=True)
df['Dependents'].value_counts().sum()
```

```python
df['Education'].value_counts()
```

```python
df['Self_Employed'].fillna('No',inplace=True)
df['Self_Employed'].value_counts()
```

```python
df['ApplicantIncome'].value_counts().sum()
```

```python
df['CoapplicantIncome'].value_counts().sum()
```

```python
df['LoanAmount'].fillna(df['LoanAmount'].median(),inplace=True)
```

```python
df['LoanAmount'].value_counts().sum()
```

```python
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median(),inplace=True)
df['Loan_Amount_Term'].value_counts().sum()
```

```python
df['Credit_History'].fillna(df['Credit_History'].median(),inplace=True)
df['Credit_History'].value_counts().sum()
```

```python
df['Property_Area'].value_counts()
```

```python
y_map={'Semiurban':0,'Urban':1,'Rural':2}
df['Property_Area']=[y_map[item] for item in df['Property_Area']]
```

```python
df['Loan_Status'].value_counts()
y_map={'Y':1,'N':0}
df['Loan_Status']=[y_map[item] for item in df['Loan_Status']]
df.dtypes
```

```python
df['Dependents'].value_counts()
y_map={'0':0,'1':1,'2':2,'3+':3}
df['Dependents']=[y_map[item] for item in df['Dependents']]
```

```python
df['Education'].value_counts()
y_map={'Graduate':0,'Not Graduate':1}
df['Education']=[y_map[item] for item in df['Education']]
```

```python
df['Self_Employed'].value_counts()
y_map={'No':0,'Yes':1}
df['Self_Employed']=[y_map[item] for item in df['Self_Employed']]
```

```python
df['ApplicantIncome'] = df['ApplicantIncome'].astype('int')
df['CoapplicantIncome']=df['CoapplicantIncome'].astype('int')
df['LoanAmount']=df['LoanAmount'].astype('int')
df['Loan_Amount_Term']=df['Loan_Amount_Term'].astype('int')
df['Credit_History']=df['Credit_History'].astype('int')
```

```python
X=df.iloc[0:,0:11]
Y=df.iloc[0:,11:12]
X
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
df=df.drop(['Loan_Status'],axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.45)
model=LogisticRegression()
model.fit(x_train,y_train.values.ravel())
predictions=model.predict(x_test)
accuracy_score(predictions,y_test)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0,test_size=0.5969355)
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train.values.ravel())
predictions=model.predict(x_test)
accuracy_score(predictions,y_test)
x_test.shape
```
